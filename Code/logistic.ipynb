```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=25, test_size=0.25)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
import pickle
from sklearn.preprocessing import StandardScaler
def standard_scaler(X_train,X_test):
    scaler=StandardScaler()
    X_train_scaler=scaler.fit_transform(X_train)
    X_test_scaler=scaler.transform(X_test)

    file=open('C:\\Users\\rampe\\Desktop\\ML_algo\\ML_algo\\Model\\standardscaler.pkl','wb')
    pickle.dump(scaler,file)
    file.close()

    return X_train_scaler,X_test_scaler
```

```python
X_train_scaler,X_test_scaler=standard_scaler(X_train,X_test)
```

```python
X_train_scaler
```

```python
from sklearn.linear_model import LogisticRegression
clf= LogisticRegression()
clf.fit(X_train_scaler, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = [{'penalty': ['l1', 'l2', 'elasticnet'], 'C':[0.25, 0.5, 0.75, 1], 'random_state': [0, 42, 50, 69, 129], 'solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}]
grid_search = GridSearchCV(estimator=clf, 
                           param_grid=parameters, 
                           scoring='accuracy', 
                           cv=10, 
                           n_jobs=-1)
grid_search.fit(X_train_scaler, y_train)
best_accuracy = grid_search.best_score_
best_parameters = grid_search.best_params_
print("Best Accuracy: {:.3f}%".format(best_accuracy*100))
print("Best Parameters:", best_parameters)
```

```python
y_pred = clf.predict(X_test_scaler)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accu = accuracy_score(y_test, y_pred)
print("Test set Accuracy: {:.3f}%".format(accu*100))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=clf, cv=10, X=X_train, y=y_train)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
from sklearn.linear_model import LogisticRegression
clf_2 = LogisticRegression(random_state=0, C=0.25, penalty='l1', solver='saga')
clf_2.fit(X_train_scaler, y_train)
```

```python
y_pred_2 = clf_2.predict(X_test)
print(y_pred_2)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=clf_2, cv=10, X=X_train, y=y_train)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard deviation: {:.2f} %".format(accuracies.std()*100))
```
